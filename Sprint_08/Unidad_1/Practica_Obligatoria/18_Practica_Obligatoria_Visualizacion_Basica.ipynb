```python
df_california = df_california.drop("Unnamed: 0", axis = 1)
df_california.info()
```

```python
media_elevation_ft = df_california["elevation_ft"].mean()
df_california["elevation_ft"] = df_california["elevation_ft"].fillna(media_elevation_ft)
df_california.info()
```

```python
media_area_total_sq_mi = df_california["area_total_sq_mi"].mean()
df_california["area_total_sq_mi"] = df_california["area_total_sq_mi"].fillna(media_area_total_sq_mi)
df_california.info()
```

```python
media_area_water_sq_mi = df_california["area_water_sq_mi"].mean()
df_california["area_water_sq_mi"] = df_california["area_water_sq_mi"].fillna(media_area_water_sq_mi)
df_california.info()
```

```python
media_area_total_km2 = df_california["area_total_km2"].mean()
df_california["area_total_km2"] = df_california["area_total_km2"].fillna(media_area_total_km2)
df_california.info()
```

```python
media_area_land_km2 = df_california["area_land_km2"].mean()
df_california["area_land_km2"] = df_california["area_land_km2"].fillna(media_area_land_km2)
df_california.info()
```

```python
media_area_water_km2 = df_california["area_water_km2"].mean()
df_california["area_water_km2"] = df_california["area_water_km2"].fillna(media_area_water_km2)
df_california.info()
```

```python
media_area_water_percent = df_california["area_water_percent"].mean()
df_california["area_water_percent"] = df_california["area_water_percent"].fillna(media_area_water_percent)
df_california.info()
```

### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
ciudades_mas_pobladas = df_california.sort_values(by = "population_total", ascending = False).head(5)

ciudades = ciudades_mas_pobladas["city"]
habitantes = ciudades_mas_pobladas["population_total"]
plt.bar(ciudades, habitantes, color = ["blue", "grey", "green", "brown"], alpha = 0.5)
plt.title("Ciudades ms pobladas de California")
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)");
```

```python
plt.hist(df_california["area_total_sq_mi"],
         bins = 50,
         alpha = 0.7,
         color = "steelblue",
         label = "Hist Area"
        )
plt.xlabel("Superficie (Km$^2$)", family = "Bradley Hand itc", size = 18)
plt.ylabel("Nmero de Ciudades", family = "Bradley Hand itc", size = 18)
plt.legend();
```

```python
plt.scatter(df_california["area_total_km2"], df_california["population_total"], color = "green")
plt.title("Poblacin por kilometro cuadrado de agua")
plt.xlabel("rea de agua en la ciudad (Km$^2$)")
plt.ylabel("Poblacin (miles de habitantes)")

props_flecha = {
    "arrowstyle" : "->"
}
plt.annotate("Outliers", xy = (1300, 3900000), xytext = (800, 3500000), \
    arrowprops = props_flecha)
plt.annotate("Outliers", xy = (965, 1360000), xytext = (800, 3500000), \
    arrowprops = props_flecha);
```

```python
plt.boxplot(df_california["elevation_m"],
            vert = False,
            patch_artist = True,
            boxprops = dict(facecolor = "red", color = "black"),
            medianprops = dict(color = "black")
            )
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.xlabel("Altitud (m)");
```

```python
fig, axs = plt.subplots(2, 2, figsize = (12, 12))

axs[0, 0].bar(ciudades, habitantes, color = ["blue", "grey", "green", "brown"], alpha=0.5)
axs[0, 0].set_title("Ciudades ms pobladas de California")
axs[0, 0].set_xlabel("Ciudades")
axs[0, 0].set_ylabel("Habitantes (miles)")

axs[0, 1].hist(df_california["area_total_sq_mi"], bins=50, alpha=0.7, color="steelblue", label="Hist Area")
axs[0, 1].set_xlabel("Superficie (Km$^2$)")
axs[0, 1].set_ylabel("Nmero de Ciudades")
axs[0, 1].legend()

axs[1, 0].scatter(df_california["area_total_km2"], df_california["population_total"], color="green")
axs[1, 0].set_title("Poblacin por kilmetro cuadrado de agua")
axs[1, 0].set_xlabel("rea de agua en la ciudad (Km$^2$)")
axs[1, 0].set_ylabel("Poblacin (miles de habitantes)")
axs[1, 0].annotate("Outliers", xy=(1300, 3900000), xytext=(800, 3500000), arrowprops=dict(arrowstyle="->"))
axs[1, 0].annotate("Outliers", xy=(965, 1360000), xytext=(800, 3500000), arrowprops=dict(arrowstyle="->"))

axs[1, 1].boxplot(df_california["elevation_m"], vert=False, patch_artist=True, boxprops=dict(facecolor="red", color="black"), medianprops=dict(color="black"))
axs[1, 1].set_title("Diagrama de caja de la Altitud de las ciudades de California")
axs[1, 1].set_xlabel("Altitud (m)");
```
